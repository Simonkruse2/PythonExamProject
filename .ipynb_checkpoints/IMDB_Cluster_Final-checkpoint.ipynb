```python
import pandas as pd 
import numpy as np

df_movie_final = pd.read_csv('movie_ratings_final.csv')
df_movie_final = df_movie_final.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
df_movie_final.loc[:, 'year'] = df_movie_final['year'].str[-5:].astype(int)
df_movie_final.loc[:, 'runtime'] = df_movie_final['runtime'].str[:-3].astype(int)
df_movie_final.head()
```

```python
# Sanitize budget for currency signs
df_movie_final = df_movie_final.drop(df_movie_final[df_movie_final['budget'].str.startswith('$') == False].index)
```

```python
import re
listbudget = df_movie_final.budget
listCurrency = []
for ele in listbudget:
    listCurrency.append(re.compile("\d").split(ele)[0])
listCurrency
df_currency = pd.DataFrame(listCurrency, columns=["currency"])
df_currency.currency.unique()
```

```python
df_movie_final.loc[:, 'budget'] = df_movie_final['budget'].str.replace("$", "").astype(int)
df_movie_final.reset_index(drop=True, inplace=True)
df_movie_final.head()
```

```python
df_genres = df_movie_final['genres'].str.replace("[", "").str.replace("]", "").str.replace("'", "").str.replace("'","").str.join(sep='').str.get_dummies(sep=', ')
result = pd.concat([df_movie_final, df_genres], axis=1).reindex(df_movie_final.index)

# sanitize ageRestriction and drop unrelated columns
result = result.drop(result[result['ageRestriction']=='Unrated'].index)
result = result.drop(result[result['ageRestriction']=='Not Rated'].index)
result = result.drop(['title','directors','actors', 'genres', 'link'],'columns')

# onehot encode ageRestriction
result = pd.get_dummies(result,columns=['ageRestriction'])
result.reset_index(drop=True, inplace=True)
result.head()
```

```python

list_profit_margin = []
for i in range(len(result)):
    list_profit_margin.append(result.gross[i] / result.budget[i])


df_profit_margin = pd.DataFrame(list_profit_margin, columns=['profit_margin'])
result = pd.concat([result, df_profit_margin], axis=1).reindex(result.index)

result.head()
```

```python
from sklearn.cluster import estimate_bandwidth

def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)

movie_data = clean_dataset(result)

eb = estimate_bandwidth(movie_data)
```

```python
from sklearn.cluster import MeanShift
analyzer = MeanShift(bandwidth=eb) 
analyzer.fit(movie_data)
```

```python
labels = analyzer.labels_
print(labels)
print('\n\n',np.unique(labels))
```

```python
movie_data['cluster_group'] = np.nan
data_length=len(movie_data)
for i in range(data_length): #movie length
    movie_data.iloc[i,movie_data.columns.get_loc('cluster_group')] = labels[i] #set the cluster label on each row

movie_data.head()
```

```python
movie_data.describe()
```

```python
#Grouping movies by Cluster
movie_cluster_data = movie_data.groupby(['cluster_group']).mean()
#Count of movies in each cluster
movie_cluster_data['Counts'] = pd.Series(movie_data.groupby(['cluster_group']).size())
movie_cluster_data
```
